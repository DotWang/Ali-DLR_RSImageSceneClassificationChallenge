```python
Data1=pd.read_csv("/data/DW/Challenge/GermanAIChallenge2018/Round_2/R2_preda_single_optionA_Res50_30_fold1_0124.csv",header=None)
Data2=pd.read_csv("/data/DW/Challenge/GermanAIChallenge2018/Round_2/R2_preda_single_optionB_Res50_50_fold1_0126.csv",header=None)
Data3=pd.read_csv("/data/DW/Challenge/GermanAIChallenge2018/Round_2/R2_preda_single_optionB_Res50_epoch40_fold1_0127.csv",header=None)
Data4=pd.read_csv("/data/DW/Challenge/GermanAIChallenge2018/Round_2/R2_preda_single_optionB_Res50_epoch30_fold2_0127.csv",header=None)
Data5=pd.read_csv("/data/DW/Challenge/GermanAIChallenge2018/Round_2/R2_preda_single_optionB_Res50_epoch40_fold3_0128.csv",header=None)
Data6=pd.read_csv("/data/DW/Challenge/GermanAIChallenge2018/Round_2/R2_preda_single_optionB_Res50_epoch55_fold4_0129.csv",header=None)
Data7=pd.read_csv("/data/DW/Challenge/GermanAIChallenge2018/Round_2/R2_preda_single_optionB_Res50_epoch45_fold5_0129.csv",header=None)
```

```python
a1=np.array(Data1)
a2=np.array(Data2)
a3=np.array(Data3)
a4=np.array(Data4)
a5=np.array(Data5)
a6=np.array(Data6)
a7=np.array(Data7)
```
