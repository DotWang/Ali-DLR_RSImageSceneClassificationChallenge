```python
Labels=Labels.tolist()
```

```python
type(Labels)
```

```python
LargeClass=[]
```

```python
for i in Labels:
    if i>=0 and i<10:
        LargeClass.append(1)
    elif i>=10 and i<14:
        LargeClass.append(2)
    elif i>=14 and i<16:
        LargeClass.append(3)
    else:
        LargeClass.append(4)
```

```python
LargeClass
```

```python
LargeClass = np.array(LargeClass)
```
